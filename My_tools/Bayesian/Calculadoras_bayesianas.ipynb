```python
import pandas as pd
import re
```

```python
df = pd.read_csv('censo.csv')
df.head()
```

```python
# Forma rápida:

# Prob ser mujer P(A=mujer)
(df['sex']=='Female').mean()

# Prob ser mujer condicionado a ser blanco P(A=Mujer/B=Blanco)
(df[df['race']=='White']['sex']=='Female').mean()
```

```python
# Lista de casos posibles para una columna
def possible_cases(df, target):
    return df[target].unique()

# Probabilidad para cada uno de los casos posibles de una columna
def describe_cases(df, target):
    cases = possible_cases(df, target)
    output = {}
    for c in cases:
        output[c] = (df[target]==c).mean()
    return output

# Probabilidad P(A=a)
def probability(df, target):
    splited = re.split('(\W)', target)
    key = splited[0]
    value = splited[-1]
    symbol = ''.join([x for x in splited if x in ['=', '>', '<']])
    string = "(df['{}']{}'{}').mean()".format(key,symbol,value)
    return eval(string)

# Probabilidad condicional P(A=a/B=b)
def conditional_probability(df, target, condition):
    splited_condition = re.split('(\W)', condition)
    key_condition = splited_condition[0]
    value_condition = splited_condition[-1]
    symbol_condition = ''.join([x for x in splited_condition if x in ['=', '>', '<']])
    string_condition = "df[df['{}']{}'{}']".format(key_condition,symbol_condition,value_condition)
    df = eval(string_condition)
    
    splited_target = re.split('(\W)', target)
    key_target = splited_target[0]
    value_target = splited_target[-1]
    symbol_target = ''.join([x for x in splited_target if x in ['=', '>', '<']])
    string_target = "(df['{}']{}'{}').mean()".format(key_target,symbol_target,value_target)
    return eval(string_target)

# Probabilidad condicional P(A=a/B=b)
def conditional_probability(df, target, condition):
    splited_condition = re.split('(\W)', condition)
    key_condition = splited_condition[0]
    value_condition = splited_condition[-1]
    symbol_condition = ''.join([x for x in splited_condition if x in ['=', '>', '<']])
    string_condition = "df[df['{}']{}'{}']".format(key_condition,symbol_condition,value_condition)
    df = eval(string_condition)
    
    splited_target = re.split('(\W)', target)
    key_target = splited_target[0]
    value_target = splited_target[-1]
    symbol_target = ''.join([x for x in splited_target if x in ['=', '>', '<']])
    string_target = "(df['{}']{}'{}').mean()".format(key_target,symbol_target,value_target)
    return eval(string_target)
```

```python
possible_cases(df, 'race')
possible_cases(df, 'sex')
describe_cases(df,'race')
```

```python
# P(A)P(B/A) = P(B)P(A/B) 
# P(A=White)P(B=Female/A=White) = P(B=Female)P(A=White/B=Female) 
right = probability(df,'race==White') * conditional_probability(df,'sex==Female','race==White')
left = probability(df,'sex==Female') * conditional_probability(df,'race==White','sex==Female')
print('P(A)P(B/A) = ' + str(left))
print('P(B)P(A/B) = ' + str(right))
print('P(A)P(B/A) = P(B)P(A/B) : ' + str(round(right,5)==round(left,5)))
```

```python
# Probabilidad total
# P(A=a) = P(A=a/B=b1)P(B=b1) +  P(A=a/B=b2)P(B=b2)

# P(A)P(B/A) = P(B)P(A/B) 
one = probability(df,'sex==Female') * conditional_probability(df,'race==White','sex==Female')
two = probability(df,'sex==Male') * conditional_probability(df,'race==White','sex==Male')
prob = probability(df,'race==White')
print('P(A=White/B=Female)P(B=Female) = ' + str(one))
print('P(A=White/B=Male)P(B=Male)     = ' + str(two))
print('P(A=White)                     = ' + str(prob))
print('P(A=a) = P(A=a/B=b1)P(B=b1) + P(A=a/B=b2)P(B=b2) : ' + str(round(prob,5)==round(one + two,5)))
```
